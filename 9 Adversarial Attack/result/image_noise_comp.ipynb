```python
import numpy as np 
import matplotlib.pyplot as plt 

from data_get import get_data, add_noise, image_compress
```

```python
# plot image 
def plot_n_img(n, x, size = 28): 
    plt.figure(figsize=(4*n, 4))
    for i in range(1, n+1): 
        ax = plt.subplot(1, n, i)
        plt.imshow(x[i-1].reshape(size, size))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()
```

```python
x_train, y_train, x_valid, y_valid, x_test, y_test  = get_data()
x_train_noise, x_valid_noise, x_test_noise = add_noise(x_train, x_valid, x_test, mean=0, sd=0)
plot_n_img(5, x_train_noise)
```
