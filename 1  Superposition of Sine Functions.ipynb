```python
import numpy as np 
from math import pi 
```

```python
# random sample from U(-pi,pi)
x = np.random.uniform(-pi,pi,512)
x.sort()
b1,b2,b3 = np.random.uniform(0,2*pi,3)
y = np.sin(x) + np.sin(10*x+b1) + np.sin(20*x+b2) + np.sin(30*x+b3)

# reshaping 
x = x.reshape((len(x), 1))
y = y.reshape((len(y), 1))
```

```python
import tensorflow as tf 
from tensorflow import keras 
```

```python
# define the keras model 
K = 64
input_layer = keras.Input(shape = (1, ))
hidden_layer_1 = keras.layers.Dense(units = K, 
                                    activation = 'sigmoid')
hidden_layer_2 = keras.layers.Dense(units = K, 
                                    activation = 'sigmoid')
hidden_layer_3 = keras.layers.Dense(units = K, 
                                    activation = 'sigmoid')
hidden_layer_4 = keras.layers.Dense(units = K, 
                                    activation = 'sigmoid')
hidden_layer_5 = keras.layers.Dense(units = K, 
                                    activation = 'sigmoid')
output_layer = keras.layers.Dense(units = 1, 
                                    use_bias = False)
model = keras.Sequential([input_layer, hidden_layer_1,
                          hidden_layer_2, hidden_layer_3, 
                          hidden_layer_4, hidden_layer_5, 
                          output_layer])

# compling the model 
adam = keras.optimizers.Adam(learning_rate=0.0005)
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
import matplotlib.pyplot as plt 
```

```python
# fitting the data 
beta_1 = model.fit(x, y, epochs=20, batch_size=32)

# predicting the data
predict_1 = model.predict(x)

# plot the graph
plt.figure('20 Epochs', figsize = (10,7))
plt.plot(x, y)
plt.plot(x, predict_1, 'r*')
plt.title('20 Epochs', fontsize=16)
plt.ylabel('y', fontsize=15)
plt.xlabel('x', fontsize=15)
plt.legend(['f(x)', 'beta(x)'], loc='upper left', fontsize=12)
plt.show()
```
